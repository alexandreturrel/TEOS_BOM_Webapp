# Getting the current Pricelist from Excel

### import libraries

```python
import pandas as pd
import string
import re
```

### get Pricelist file

```python
pricelist = 'pricelist may 2020.xlsx'
output =  'out.xlsx'
```

### reading the Pricelist and getting only Solutions & Professional Displays references

```python
df_s = pd.read_excel(pricelist, sheet_name="Solutions")
df_d = pd.read_excel(pricelist, sheet_name="Professional Displays")
for i in range(df_s.columns.size):
    df_s.rename(columns={ df_s.columns[i]: string.ascii_letters[i] }, inplace = True)
for i in range(df_d.columns.size):
    df_d.rename(columns={ df_d.columns[i]: string.ascii_letters[i] }, inplace = True)
```

### renaming columns to edit content and filter

#### Solutions

```python
df_s.rename(columns={
    "a": "Region",
    "c": "Tag",
    "e": "Month", 
    "f": "Category",
    "h": "Reference",
    "i": "Description",
    "m": "Buy Price",
    "r": "List Price",
    "s": "Comments"
}, inplace=True)

tmp = df_s['Tag']
tmp = tmp.str.split('|')
tmp = tmp.str[1]
df_s['Tag'] = tmp
```

##### dropping Education Solutions

```python
df_s = df_s.loc[(df_s['Month'] == 'May_2020') & (df_s['Category']!='UCB')]

df_s = df_s.loc[(df_s['Tag'] != 'VisionExchange')
           & (df_s['Tag'] != 'VisionExchangeSoftware')
           & (df_s['Tag'] != 'VisionExchangeTriallicenses')
           & (df_s['Tag'] != 'VisionExchangeSupport')
           & (df_s['Tag'] != 'VisionPresenterWarrantyExtension')
           & (df_s['Tag'] != 'EducationSolutionsTraining')
           & (df_s['Tag'] != 'VisonExchange')
           & (df_s['Tag'] != 'VisonExchangeSoftware')
           & (df_s['Tag'] != 'VisonExchangeSupport')
           & (df_s['Tag'] != 'VisonPresenterWarrantyExtension')]

df_s.reset_index(drop=True, inplace=True)
```

#### Displays

```python
df_d.rename(columns={
    "a": "Region",
    "c": "Tag",
    "e": "Month", 
    "f": "Category",
    "h": "Reference",
    "x": "Buy Price",
    "C": "List Price",
    "D": "Comments"
}, inplace=True)

tmp = df_d['Tag']
tmp = tmp.str.split('|')
tmp = tmp.str[1]
df_d['Tag'] = tmp
```

##### filtering only (insert next title) references lines to rename correctly valid columns

Displays

```python
df_display = df_d.copy()
df_display = df_display.loc[(df_d['Month'] == 'May_2020') & (df_d['Tag']=='Displays')]
df_display.rename(columns={
    "i": "Size",
    "j": "Technology",
    "k": "Resolution",
    "l": "HDR",
    "m": "Brightness",
    "n": "24/7",
    "o": "Portrait",
    "p": "Tuner",
    "q": "IP Control",
    "r": "RS232",
    "s": "Professional Mode",
    "t": "Warranty",
}, inplace=True)

df_d = df_d.loc[(df_d['Tag'] != 'Displays')]
df_d.reset_index(drop=True, inplace=True)
```

Touch Overlays

```python
df_to = df_d.copy()
df_to = df_to.loc[(df_d['Month'] == 'May_2020') & (df_d['Tag']=='TouchOverlays')]
df_to.rename(columns={
    "i": "Size",
    "j": "Technology",
    "k": "Interactive",
    "l": "Color",
    "m": "Compatible Models",
}, inplace=True)

df_d = df_d.loc[(df_d['Tag'] != 'TouchOverlays')]
df_d.reset_index(drop=True, inplace=True)
```

Totem & Kiosks

```python
df_tk = df_d.copy()
df_tk = df_tk.loc[(df_d['Month'] == 'May_2020') & (df_d['Tag']=='Totemskiosks')]
df_tk.rename(columns={
    "i": "Size",
    "j": "Technology",
    "k": "Interactive",
    "l": "Color",
    "m": "Compatible Models",
}, inplace=True)

df_d = df_d.loc[(df_d['Tag'] != 'Totemskiosks')]
df_d.reset_index(drop=True, inplace=True)
```

```python
df_d = pd.concat([df_d, df_display, df_to, df_tk])
```

```python
df_d = df_d.loc[(df_d['Month'] == 'May_2020')]
df_d.rename(columns={
    "i": "Description",
}, inplace=True)
```

```python
df_d.reset_index()
```

```python
df = pd.concat([df_s, df_d])
```

correcting Tags

```python
df.loc[df['Description'].str.contains('PrimeSupportPro extension') 
       & (df['Tag'] != 'WarrantyExtension'),
      ['Tag']] = 'WarrantyExtension'

df.loc[df['Reference'].str.startswith('TEM'), ['Tag']] = 'TEOSLicence'
df.loc[df['Reference'].str.startswith('TDM'), ['Tag']] = 'TDMLicence'
```

```python
for column_name in df.columns:
    if column_name in string.ascii_letters:
        df.drop(column_name, axis=1, inplace=True)
```

```python
df['Region']='EU'
```

# exporting new file

```python
df.to_excel(output, index=False)
```
